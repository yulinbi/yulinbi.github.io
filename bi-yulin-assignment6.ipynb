```python
import bqplot
import numpy as np
import ipywidgets
import pandas as pd
import random
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import Javascript, display, clear_output
import ipywidgets as widgets
```

```python
#load original data
data = pd.read_csv('https://uiuc-ischool-dataviz.github.io/is445_AOUAOG_fall2021/week02/data/building_inventory.csv')
```

```python
#show data
data
```

```python
#dealing with yeardata 
data = data[data['Year Constructed']!=0]
data = data[data['Year Acquired']!=0]
```

```python
#dealing with countydata and agencydata
data = data.fillna({'Agency Name':'unknown', 'County': 'unknown'})
data
```

```python
# Lets down sample to 100 samples in order to show a beautiful heatmap
nsamples = 100
downSampleMask = np.random.choice(range(len(data)-1), nsamples, replace=False)
datas = data.loc[downSampleMask]
datas = datas[['Agency Name','County','Square Footage','Year Acquired']]
datas.dropna(subset=['Agency Name','County','Square Footage','Year Acquired'])
datas
```

```python
#calculate the average 
avg = datas.groupby(['Agency Name','County'])['Square Footage'].mean()
avg
```

```python
#transfer the data
df=pd.DataFrame(avg)
df.columns=['Average Square Footage']
newdata=df.reset_index() 

newdata
```

```python
#transfer the data
temp=newdata.pivot(index='Agency Name',columns='County',values='Average Square Footage')
temp
```

```python
#creat the plots
#1 create label
#NOTE: although I draw the heatmap use sample, showing barplot I use the original dataset
mySelectedLabel = ipywidgets.Label(value='Total Square Footage in all agency and county') 


#2 create the function to connect heatmap and barplot
def get_data_value(change):
    if len(change['owner'].selected) == 1: #only 1 selected
        
        item = []
        agencyindex = change['owner'].selected[0][0] #agency
        countycolumn = change['owner'].selected[0][1] #county
        item.append(agencyindex)
        item.append(countycolumn)
        agencyname = temp.index[item[0]]
        countyname = temp.columns[item[1]]
        
        mySelectedLabel.value = 'Total Square Footage of ' + agencyname + ' in ' + countyname # set our label
        
        pltdata = data[(data['County']==countyname)&(data['Agency Name']==agencyname)][['County','Agency Name','Year Acquired','Square Footage']]
        pltdata = pltdata.groupby('Year Acquired')['Square Footage'].sum()
        df=pd.DataFrame(pltdata)
        df.columns=['Total Square Footage'] 
        pltdata=df.reset_index() 
        year = pltdata['Year Acquired']
        square = pltdata['Total Square Footage']
        barplot.x = year
        barplot.y = square


        

#3 HeatMap
##1 create heatmap scale
col_sc = bqplot.ColorScale()
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()

##2 create heatmap axis
c_ax = bqplot.ColorAxis(scale = col_sc, 
                        orientation = 'vertical', 
                        side = 'right',tick_style={'font-size':7})
x_ax = bqplot.Axis(scale = x_sc,label = 'County',tick_rotate=90,tick_style={'font-size':8})
y_ax = bqplot.Axis(scale = y_sc, 
                   label = 'Agency Name',
                   orientation = 'vertical',tick_rotate=45,tick_style={'font-size':8})

##3 create heatmap mark
heat_map = bqplot.GridHeatMap(color = temp,
                              row = temp.index,
                              column = temp.columns,
                              scales = {'color': col_sc,
                                        'row': y_sc,
                                        'column': x_sc},
                              interactions = {'click': 'select'}, # make interactive on click of each box
                              anchor_style = {'fill':'blue'}, # to make our selection blue
                              selected_style = {'opacity': 1.0}, # make 100% opaque if box is selected
                              unselected_style = {'opacity': 0.8}) # make a little see-through if not



#4 BarPlot
##1 create barplot scale
x_scl = bqplot.LinearScale() 
y_scl = bqplot.LinearScale()

##2 create barplot axis
ax_xcl = bqplot.Axis(label='Year Aquired', scale=x_scl)
ax_ycl = bqplot.Axis(label='Total Square footage', scale=y_scl, 
                    orientation='vertical', side='left')

##3 create default barplot mark
initdata = data[['County','Agency Name','Year Acquired','Square Footage']]
initdata = initdata.groupby('Year Acquired')['Square Footage'].sum()
df=pd.DataFrame(initdata)
df.columns=['Total Square Footage'] 
initdata=df.reset_index() 
year = initdata['Year Acquired']
square = initdata['Total Square Footage']

barplot = bqplot.Bars(x=year,y=square,
                           scales={'x':x_scl, 'y':y_scl})


```

```python
#5 connect heatmap and barplot
heat_map.observe(get_data_value, 'selected')


#6 create figures

fig_heatmap = bqplot.Figure(marks = [heat_map], 
                    axes = [c_ax, y_ax, x_ax],title='HeatMap')


fig_barplot = bqplot.Figure(marks=[barplot],axes=[ax_xcl,ax_ycl],title='BarPlot')

```

```python
#7 make dashboard
fig_heatmap.layout.min_width='500px'
fig_barplot.layout.min_width='500px'

plots = ipywidgets.VBox([mySelectedLabel,fig_barplot])
myDashboard = ipywidgets.HBox([fig_heatmap,plots])
myDashboard
```

## Any data transformations or rescalings you did for your plot
### 1) Sampling: The original dataset is too big, so the heatmap maybe not that clear. I just select 100 rows randomly from the original data to draw the heatmap;

### 3) Pivoting: In order to deal with the error 'tuple index out of range' and match the distribution of heatmap, I use pivot() function to transfer the averaged dataset ---- The first two columns "County" and "Agency Name" are changed to the column and index; the 'Average Square Footage' column is changed to the value.

### Regarding to the year data, I select the those yeardata not equal to 0 from the original dataset;
### Regarding to the county and agency data, I use 'unknown' to fill the nan value.
